```python

import ast


def evaluation_detection(data_path):
        
    base = [json.loads(q) for q in open(data_path, "r")]
    correct=0
    incorrect=0

    for i in range(len(base)):
        test_pred = base[i]["text"]
        test_label = base[i]["ground_truth"]
        if test_pred == "None":
            if test_label == "None":
                correct += 1
        else:
            if test_label == "None":
                continue
            else:
                try:
                    test_pred = ast.literal_eval(test_pred)
                    test_label = ast.literal_eval(test_label)

                    # calculate IOU between every pred and label
                    count = 0
                    for pred in test_pred:
                        IOU = []
                        for label in test_label:
                            pred_x1, pred_y1, pred_x2, pred_y2 = pred
                            label_x1, label_y1, label_x2, label_y2 = label
                            x1, y1 = max(pred_x1, label_x1), max(pred_y1, label_y1)
                            x2, y2 = min(pred_x2, label_x2), min(pred_y2, label_y2)
                            intersection = max(0, x2 - x1) * max(0, y2 - y1)
                            union = (pred_x2 - pred_x1) * (pred_y2 - pred_y1) + (label_x2 - label_x1) * (label_y2 - label_y1) - intersection
                            IOU.append(intersection / union)
                        if max(IOU) >= 0.25:
                            count += 1   
                    if count == len(test_label):
                        correct += 1
                except:
                    continue
    print(f"obj_detection test accuracy: {correct / len(base)}")
'''
evaluation_detection("/mnt/public/vila-1.5/our_results/sv/ObjectDetection/vila_train_sisv_sftV2.jsonl")
evaluation_detection("/mnt/public/vila-1.5/our_results/sv/ObjectDetection/vila_train_sisv_pretrain_35w.jsonl")
evaluation_detection("/mnt/public/vila-1.5/our_results/sv/ObjectDetection/vila_train_sisv_pretrain_35wV2.jsonl")
evaluation_detection("/mnt/public/vila-1.5/our_results/sv/ObjectDetection/vila_train_sisv_pretrain_35w_att.jsonl")
evaluation_detection("/mnt/public/vila-1.5/our_results/sv/ObjectDetection/vila_train_sisv_pretrain_35w_attV2.jsonl")

evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_att_186w_clean.jsonl")

evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_clean_8b.jsonl")
evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_186w_clean.jsonl")

evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_clean_8b.jsonl")
evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_186w_clean_only_si.jsonl")
evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_186w_clean_only_sv.jsonl")
evaluation_detection("/mnt/public/vila-1.5/paper_results/sv/ObjectDetection/vila_train_all_sisv_att_186w_clean_13bV2.jsonl")
'''
evaluation_detection("/data3/zhangxin/wuwen/vila-1.5/rebuttal_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_att_35w_clean_8bV4.jsonl")
evaluation_detection("/data3/zhangxin/wuwen/vila-1.5/rebuttal_results/sv/ObjectDetection/Llama-3-VILA1.5-8B.jsonl")
evaluation_detection("/data3/zhangxin/wuwen/vila-1.5/last_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_att_sharegpt4v_186w_clean_8B_llava_instruct150k.jsonl")

evaluation_detection("/data3/zhangxin/wuwen/vila-1.5/last_results/sv/ObjectDetection/vila_train_all_sisv_pretrain_att_186w_clean_8B_llava_instruct150k.jsonl")
```
